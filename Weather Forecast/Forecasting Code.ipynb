```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

# Load Dataset
df = pd.read_csv("Weather Data.csv")

# Display first 5 rows
print(df.head())
```

```python
# --------------------------------------------------
# 1. FIND UNIQUE WINDSPEED
# --------------------------------------------------
print("\n1. Unique Wind Speeds:")
print(df['Wind Speed_km/h'].unique())
print(df["Wind Speed_km/h"].value_counts()) 
```

```python
# --------------------------------------------------
# 2. FIND THE NUMBER OF TIMES WHEN WEATHER IS EXACTLY CLEAR
# --------------------------------------------------
print("\n2. Number of times weather is Clear:")
print(df[df['Weather'] == 'Clear'].shape[0])
```

```python
# --------------------------------------------------
# 3. FIND THE WINDSPEED IS EXACTLY 4
# --------------------------------------------------
print("\n3. Records where Wind Speed = 4:")
print(df[df['Wind Speed_km/h'] == 4])
```

```python
# --------------------------------------------------
# 4. FIND THE MEAN VISIBILITY
# --------------------------------------------------
print("\n4. Mean Visibility:", df['Visibility_km'].mean())
```

```python
# --------------------------------------------------
# 5. FIND THE STANDARD DEVIATION OF PRESSURE KPA
# --------------------------------------------------
print("\n5. Standard Deviation of Pressure:")
print(df['Press_kPa'].std())
```

```python
# --------------------------------------------------
# 6. FIND ALL INSTANCES WHEN SNOW WAS RECORDED
# --------------------------------------------------
print("\n6. Records where Snow was recorded:")
print(df[df['Weather'].str.contains('Snow', case=False)])
```

```python
# --------------------------------------------------
# 7. SHOW ALL RECORDS WHEN WEATHER CONDITION IS FOG
# --------------------------------------------------
print("\n7. Records where Weather = Fog:")
print(df[df['Weather'] == 'Fog'])
```

```python
# --------------------------------------------------
# 8. FIND MEAN VALUE OF EACH COLUMN AGAINST EACH WEATHER CONDITION
# --------------------------------------------------
print("\n8. Mean values grouped by Weather Condition:")
print(df.groupby('Weather').mean(numeric_only=True))
```

```python
# --------------------------------------------------
# 9. FIND ALL INSTANCES WHERE WEATHER IS CLEAR
#    OR VISIBILITY IS ABOVE 40
# --------------------------------------------------
print("\n9. Weather Clear OR Visibility > 40:")
print(df[(df['Weather'] == 'Clear') | (df['Visibility_km'] > 40)])
```

```python
# --------------------------------------------------
# 10. FIND ALL INSTANCES WHERE WEATHER IS CLEAR
#     AND REL HUM_% IS ABOVE 50
# --------------------------------------------------
print("\n10. Weather Clear AND Rel Hum_% > 50:")
print(df[(df['Weather'] == 'Clear') & (df['Rel Hum_%'] > 50)])
```

```python
# Graph 1: Weather Condition Count
plt.figure(figsize=(10,15))
sns.countplot(y='Weather', data=df)
plt.title("Weather Conditions Count")
plt.xlabel("Count")
plt.ylabel("Weather")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.hist(df["Temp_C"], bins=20)
plt.title("Temperature Distribution")
plt.xlabel("Temperature (°C)")
plt.ylabel("Frequency")
plt.show()
```

```python
plt.figure(figsize=(12,5))
plt.plot(df["Temp_C"])
plt.title("Temperature Trend")
plt.xlabel("Record Number")
plt.ylabel("Temperature (°C)")
plt.show()
```

```python
plt.figure(figsize=(8,5))
plt.scatter(df["Temp_C"], df["Rel Hum_%"])
plt.title("Temperature vs Relative Humidity")
plt.xlabel("Temperature (°C)")
plt.ylabel("Relative Humidity (%)")
plt.show()
```

```python
weather = df["Weather"].value_counts().head(10)

plt.figure(figsize=(8,8))
plt.pie(weather, labels=weather.index, autopct="%1.1f%%")
plt.title("Top 10 Weather Conditions")
plt.show()
```

```python
avg_temp = df.groupby("Weather")["Temp_C"].mean().sort_values()

plt.figure(figsize=(12,6))
avg_temp.plot(kind="bar")
plt.title("Average Temperature by Weather")
plt.xlabel("Weather")
plt.ylabel("Average Temperature")
plt.xticks(rotation=90)
plt.show()
```

```python
plt.figure(figsize=(6,5))
plt.boxplot(df["Wind Speed_km/h"])
plt.title("Wind Speed Distribution")
plt.ylabel("Wind Speed (km/h)")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

plt.figure(figsize=(8,6))

sns.heatmap(
    df[["Temp_C","Dew Point Temp_C","Rel Hum_%","Wind Speed_km/h","Visibility_km","Press_kPa"]].corr(),
    annot=True,
    cmap="coolwarm"
)

plt.title("Correlation Heatmap")
plt.show()
```
